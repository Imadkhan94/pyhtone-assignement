```python
Release_mean=int(data["Release Date"].mean())
Release_mean
```

```python
def new_values(val):
    if val < Release_mean:
        return "Old movies";
    elif val >= Release_mean:
        return "New movies";
data["new/old movies"]=data["Release Date"].apply(new_values)    
```
