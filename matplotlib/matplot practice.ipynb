```python
import matplotlib.pyplot as  plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),"--r",label="sin(x)")
plt.plot(x,np.cos(x),":y",label="cos(x)")
plt.legend()
plt.show()
```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),"--r",label="sin(x)")
plt.plot(x,np.cos(x),":y",label="cos(x)")
plt.legend(loc="lower right")
plt.show()
```

```python
x=np.linspace(0,10,100)
plt.subplot(3,1,1)
plt.plot(x,np.sin(x))
plt.subplot(3,1,2)
plt.plot(x,np.cos(x))
plt.subplot(3,1,3)
plt.plot(x,np.tan(x))
plt.show()
```

```python
x=np.linspace(0,10,100)
plt.subplot(3,1,1)
plt.plot(x,np.sin(x),label="sin")
plt.legend()
plt.subplot(3,1,2)
plt.plot(x,np.cos(x),label="cos")
plt.legend()
plt.subplot(3,1,3)
plt.plot(x,np.tan(x),"o",label="tan")
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn-whitegrid")

x=np.linspace(0,10,100)
plt.subplot(3,1,1)
plt.plot(x,np.sin(x),label="sin")
plt.legend()
plt.subplot(3,1,2)
plt.plot(x,np.cos(x),label="cos")
plt.legend()
plt.subplot(3,1,3)
plt.plot(x,np.tan(x),"o",label="tan")
plt.legend()

plt.show()
```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),"--r",label="sin(x)")
plt.plot(x,np.cos(x),":y",label="cos(x)")
plt.legend(loc="lower right")
plt.title("multiple graph")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## scatter plot

```python
x=np.linspace(0,10,30)
plt.plot(x,np.sin(x),"o",color="red") 
plt.show()
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
plt.scatter(x,y)
plt.show()
```

```python
x=np.random.randn(100)
y=np.random.randn(100)
colors=np.random.rand(100)
sizes=1000*np.random.rand(100)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.3)
plt.colorbar()
plt.show()
```

```python
data=np.random.randn(100)
plt.hist(data,bins=20,histtype="stepfilled",color="green")
plt.show()
```

```python
x=np.linspace(0,10,30)
plt.plot(x,np.sin(x),"--r",label="sin")
plt.plot(x,np.cos(x),":b",label="cos")
plt.legend(loc="upper right")
plt.show()
```
